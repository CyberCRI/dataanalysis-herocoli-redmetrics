```python
#players_stats = pd.DataFrame.from_csv("data/players_stats.csv")
players_coordinates = pd.DataFrame.from_csv("data/players_coordinates.csv")
```

```python
players_coordinates.head(5)
```

```python
logs = players_coordinates[ (players_coordinates["level"] == "sandbox2") & (players_coordinates["y"] < 1500) ] 
```

## just plot it!

```python
plt.figure(figsize=(15,15))
plt.axis('equal')

types = ["death", "reach", "equip", "craft"]
c = ["black", "red", "blue", "green"]
a = [ 0.3, 1, 1,1 ]

for i, t in enumerate(types):
  pts = logs[logs["type"] == t]
  x = pts["x"]
  y = pts["y"]

  plt.plot(x, y, 'o', c=c[i], lw=0, alpha=a[i])

plt.show()
```

```python
import matplotlib.image as image

plt.figure(figsize=(15,15))
plt.axis('equal')

im = image.imread('images/map.sandbox.png')

types = ["death", "reach", "equip", "craft"]
c = ["black", "red", "blue", "green"]
a = [ 0.3, 1, 1,1 ]

ox = -558
oy = -98
r = 0.488
plt.imshow(im, aspect='auto', zorder=-1, extent=(r * (0+ox), r * (1074+ox), r * (0+oy), r * (1023+oy)), alpha=0.5)

for i, t in enumerate(types):
  pts = logs[logs["type"] == t]
  x = pts["x"]
  y = pts["y"]

  plt.plot(x, y, 'o', c=c[i], lw=0, alpha=a[i])

plt.xlim([-210, 200])
plt.ylim([-10, 400])

plt.show()
```

## superposition of paths

```python
df = logs.loc[[ x for x in players_coordinates.index.unique() if len(players_coordinates.loc[x,:]) > 5],:]

players_list = list(df.index.unique())
plt.figure(figsize=(15,15))

ox = -558
oy = -98
r = 0.488
plt.imshow(im, aspect='auto', zorder=-1, extent=(r * (0+ox), r * (1074+ox), r * (0+oy), r * (1023+oy)), alpha=0.5)

for i, p in enumerate(players_list):
  select = df.loc[p,:]

  x = select["x"]
  y = select["y"]
  
  plt.plot(x,y, "-", c="black", alpha=0.15)

plt.axis('equal')

plt.xlim([-210, 200])
plt.ylim([-10, 400])
  
plt.show();
```

## indivual paths

```python
df = logs.loc[[ x for x in logs.index.unique() if len(logs.loc[x,:]) > 5],:]

players_list = list(df.index.unique())
plt.figure()

f, axarr = plt.subplots(len(players_list)/5+1, 5, figsize=(20,10))

for i, p in enumerate(players_list):
  select = df.loc[p,:]

  ax = axarr[i/5, i%5]

  x = select["x"]
  y = select["y"]
  
  ax.plot(x,y, "-o")
  ax.axis('equal')

  ax.set_xlim([-200, 200])
  ax.set_ylim([0, 400])

plt.show();
```
